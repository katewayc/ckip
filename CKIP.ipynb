```python
# 安裝必要的套件
# !pip install ckip-transformers pandas

from ckip_transformers.nlp import CkipWordSegmenter
import pandas as pd

# 初始化分詞器
ws_driver = CkipWordSegmenter(model="bert-base")

# 加載原始 CSV 文件
input_file = "reviews.csv"  # 替換為您的 CSV 文件名稱
output_file = "reviews_segmented.csv"  # 分詞後輸出的文件名稱

# 讀取 CSV 文件
reviews_df = pd.read_csv(input_file)

# 確保數據中的評論為字符串
reviews_df['清理後評論內容'] = reviews_df['清理後評論內容'].astype(str)

# 使用 CKIP Transformer 進行分詞
def segment_text(text):
    # 單條記錄的分詞
    segments = ws_driver([text])[0]
    return " ".join(segments)

# 對每一條評論進行分詞
reviews_df['分詞結果'] = reviews_df['清理後評論內容'].apply(segment_text)

# 保存分詞結果到新的 CSV 文件
reviews_df.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"分詞完成，結果已保存到 {output_file}")
```
